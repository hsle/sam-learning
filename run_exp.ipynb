```python
from sam_learning.sam_learning import get_precondition_effect
```

```python
npuzzle = parse_model('sam_learning/fama_learning/dataset/npuzzle/empty')
npuzzle_true = parse_model('sam_learning/fama_learning/dataset/npuzzle/reference')
```

```python
num_observations = 1

for i in range(10):
    print('------------------------------ run %s---------------------------' % str(i+1))
    random.seed(datetime.now())
    ind = random.randint(0,9)

    trace = parse_trajectory('sam_learning/fama_learning/dataset/npuzzle/trajectory-%s' % str(ind).zfill(2), npuzzle_true) 
    trace.states = trace.states[:2]
    trace.states[1].next_action = None
    trace.length = len(trace.states)
    npuzzle = parse_model('sam_learning/fama_learning/dataset/npuzzle/empty')
    get_precondition_effect(npuzzle, [trace])
    print(npuzzle)
    
```

```python
blocksworld_true = parse_model('sam_learning/fama_learning/dataset/blocks/reference')
```

```python
num_observations = 1
num_runs = 10

for i in range(num_runs):
    print('------------------------------ RUN %s ----------------------------' %str(i+1))
    random.seed(datetime.now())
    ind = random.randint(0,9)
    
    print(ind)
    blocksworld = parse_model('sam_learning/fama_learning/dataset/blocks/empty')
    
    for num_trans in range(10):
        trace = parse_trajectory('sam_learning/fama_learning/dataset/blocks/trajectory-%s' % str(ind).zfill(2), blocksworld_true)
        num_trans += 1
        print('no. transitions %s' %str(num_trans))
       
        trace.states = trace.states[:(num_trans+1)]
        trace.states[num_trans].next_action = None
        trace.length = len(trace.states)
        get_precondition_effect(blocksworld, [trace])
        print(blocksworld)
    
```
